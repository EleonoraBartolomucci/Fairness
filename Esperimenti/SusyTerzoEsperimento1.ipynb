```
from pydrive.auth import GoogleAuth
from pydrive.drive import GoogleDrive
from oauth2client.client import GoogleCredentials
from google.colab import auth
import json
import pandas as pd
import csv
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

auth.authenticate_user()
gauth = GoogleAuth()
gauth.credentials = GoogleCredentials.get_application_default()
drive = GoogleDrive(gauth)

# file con i dati di tutti i ristoranti random
sort = 'Random'

relevance = '1n9vdCnppJGdE6M6HdKYlksuTntQYHA7P'  # FILE ID, got on google drive with condivision link
download = drive.CreateFile({'id': relevance})
download.GetContentFile('AllRestaurant_' + '%s' % sort + '_Analysis.csv')

# leggo il csv con pandas
tot = pd.read_csv('AllRestaurant_' + '%s' % sort + '_Analysis.csv')
```

```
#RACCOLTA DATI

#Percentuali di maschi e femmine sopra e sotto i 40, per ogni ristorante

# converto le stringhe di age in int
tot.age = pd.to_numeric(tot.age, errors='coerce').fillna(0).astype(np.int64)

res = tot[(tot['ethnicity'] != 'Not recognized')] #assumo che i non riconosciuti siano equamente distribuiti

#NORTH CAROLINA
totaleNC = len(res[(res['zone'] == 'NC')])
NCwhite = (len(res[(res['zone'] == 'NC') & (res['ethnicity'] == 'white')]) * 100) / totaleNC
NCblack = (len(res[(res['zone'] == 'NC') & (res['ethnicity'] == 'black or african american')]) * 100) / totaleNC
NCmexican = (len(res[(res['zone'] == 'NC') & (res['ethnicity'] == 'hispanic, latino, or spanish origin')]) * 100) / totaleNC
NCasian = (len(res[(res['zone'] == 'NC') & (res['ethnicity'] == 'asian')]) * 100) / totaleNC
NCother = (len(res[(res['zone'] == 'NC') & ((res['ethnicity'] == 'american indian or alaska native') | (res['ethnicity'] == 'middle eastern or north african') | (res['ethnicity'] == 'native hawaiian or pacific islander'))]) * 100) / totaleNC
#maschi e femmine
NCwhiteM = (len(res[(res['zone'] == 'NC') & (res['ethnicity'] == 'white') & (res['gender'] == 'masculine')]) * 100) / totaleNC
NCwhiteF = (len(res[(res['zone'] == 'NC') & (res['ethnicity'] == 'white') & (res['gender'] == 'feminine')]) * 100) / totaleNC
NCblackM = (len(res[(res['zone'] == 'NC') & (res['ethnicity'] == 'black or african american') & (res['gender'] == 'masculine')]) * 100) / totaleNC
NCblackF = (len(res[(res['zone'] == 'NC') & (res['ethnicity'] == 'black or african american') & (res['gender'] == 'feminine')]) * 100) / totaleNC
NCmexicanM = (len(res[(res['zone'] == 'NC') & (res['ethnicity'] == 'hispanic, latino, or spanish origin') & (res['gender'] == 'masculine')]) * 100) / totaleNC
NCmexicanF = (len(res[(res['zone'] == 'NC') & (res['ethnicity'] == 'hispanic, latino, or spanish origin') & (res['gender'] == 'feminine')]) * 100) / totaleNC
NCasianM = (len(res[(res['zone'] == 'NC') & (res['ethnicity'] == 'asian') & (res['gender'] == 'masculine')]) * 100) / totaleNC
NCasianF = (len(res[(res['zone'] == 'NC') & (res['ethnicity'] == 'asian') & (res['gender'] == 'feminine')]) * 100) / totaleNC


#NEVADA
totaleNV = len(res[(res['zone'] == 'NV')])
NVwhite = (len(res[(res['zone'] == 'NV') & (res['ethnicity'] == 'white')]) * 100) / totaleNV
NVblack = (len(res[(res['zone'] == 'NV') & (res['ethnicity'] == 'black or african american')]) * 100) / totaleNV
NVmexican = (len(res[(res['zone'] == 'NV') & (res['ethnicity'] == 'hispanic, latino, or spanish origin')]) * 100) / totaleNV
NVasian = (len(res[(res['zone'] == 'NV') & (res['ethnicity'] == 'asian')]) * 100) / totaleNV
NVother = (len(res[(res['zone'] == 'NV') & ((res['ethnicity'] == 'american indian or alaska native') | (res['ethnicity'] == 'middle eastern or north african') | (res['ethnicity'] == 'native hawaiian or pacific islander'))]) * 100) / totaleNV
#maschi e femmine
NVwhiteM = (len(res[(res['zone'] == 'NV') & (res['ethnicity'] == 'white') & (res['gender'] == 'masculine')]) * 100) / totaleNV
NVwhiteF = (len(res[(res['zone'] == 'NV') & (res['ethnicity'] == 'white') & (res['gender'] == 'feminine')]) * 100) / totaleNV
NVblackM = (len(res[(res['zone'] == 'NV') & (res['ethnicity'] == 'black or african american') & (res['gender'] == 'masculine')]) * 100) / totaleNV
NVblackF = (len(res[(res['zone'] == 'NV') & (res['ethnicity'] == 'black or african american') & (res['gender'] == 'feminine')]) * 100) / totaleNV
NVmexicanM = (len(res[(res['zone'] == 'NV') & (res['ethnicity'] == 'hispanic, latino, or spanish origin') & (res['gender'] == 'masculine')]) * 100) / totaleNV
NVmexicanF = (len(res[(res['zone'] == 'NV') & (res['ethnicity'] == 'hispanic, latino, or spanish origin') & (res['gender'] == 'feminine')]) * 100) / totaleNV
NVasianM = (len(res[(res['zone'] == 'NV') & (res['ethnicity'] == 'asian') & (res['gender'] == 'masculine')]) * 100) / totaleNV
NVasianF = (len(res[(res['zone'] == 'NV') & (res['ethnicity'] == 'asian') & (res['gender'] == 'feminine')]) * 100) / totaleNV


#ARIZONA
totaleAZ = len(res[(res['zone'] == 'AZ')])
AZwhite = (len(res[(res['zone'] == 'AZ') & (res['ethnicity'] == 'white')]) * 100) / totaleAZ
AZblack = (len(res[(res['zone'] == 'AZ') & (res['ethnicity'] == 'black or african american')]) * 100) / totaleAZ
AZmexican = (len(res[(res['zone'] == 'AZ') & (res['ethnicity'] == 'hispanic, latino, or spanish origin')]) * 100) / totaleAZ
AZasian = (len(res[(res['zone'] == 'AZ') & (res['ethnicity'] == 'asian')]) * 100) / totaleAZ
AZother = (len(res[(res['zone'] == 'AZ') & ((res['ethnicity'] == 'american indian or alaska native') | (res['ethnicity'] == 'middle eastern or north african') | (res['ethnicity'] == 'native hawaiian or pacific islander'))]) * 100) / totaleAZ
#maschi e femmine
AZwhiteM = (len(res[(res['zone'] == 'AZ') & (res['ethnicity'] == 'white') & (res['gender'] == 'masculine')]) * 100) / totaleAZ
AZwhiteF = (len(res[(res['zone'] == 'AZ') & (res['ethnicity'] == 'white') & (res['gender'] == 'feminine')]) * 100) / totaleAZ
AZblackM = (len(res[(res['zone'] == 'AZ') & (res['ethnicity'] == 'black or african american') & (res['gender'] == 'masculine')]) * 100) / totaleAZ
AZblackF = (len(res[(res['zone'] == 'AZ') & (res['ethnicity'] == 'black or african american') & (res['gender'] == 'feminine')]) * 100) / totaleAZ
AZmexicanM = (len(res[(res['zone'] == 'AZ') & (res['ethnicity'] == 'hispanic, latino, or spanish origin') & (res['gender'] == 'masculine')]) * 100) / totaleAZ
AZmexicanF = (len(res[(res['zone'] == 'AZ') & (res['ethnicity'] == 'hispanic, latino, or spanish origin') & (res['gender'] == 'feminine')]) * 100) / totaleAZ
AZasianM = (len(res[(res['zone'] == 'AZ') & (res['ethnicity'] == 'asian') & (res['gender'] == 'masculine')]) * 100) / totaleAZ
AZasianF = (len(res[(res['zone'] == 'AZ') & (res['ethnicity'] == 'asian') & (res['gender'] == 'feminine')]) * 100) / totaleAZ
```

```
#NORD CAROLINA YELP
labelsInside = ['White','Black or african american', 'Hispanic, latino, or spanish origin', 'Asian', 'Other']
sizesInside = [NCwhite, NCblack, NCmexican, NCasian, NCother] 

labelsOutside = ['♀', '♂', '♀', '♂', '♀', '♂', '♀', '♂', '♂ & ♀']
sizesOutside = [NCwhiteF, NCwhiteM, NCblackF, NCblackM, NCmexicanF, NCmexicanM, NCasianF, NCasianM, NCother] 

matplotlib.rcParams['figure.figsize'] = (8,8)

plt.pie(sizesOutside,  radius=0.95, startangle = 0,
        colors=['mediumorchid',  'violet',#'xkcd:baby pink',
                'limegreen', 'greenyellow',
                'darkorange', 'lightsalmon',
                'turquoise', 'paleturquoise',
                'gold'],
        wedgeprops=dict(width=0.25, edgecolor='white'),
        labels=labelsOutside, autopct='%.2f%%', textprops={'fontsize': 11},
        pctdistance=0.85, labeldistance = 1.1)

plt.pie(sizesInside, radius=0.7, startangle = 0,
        colors=['darkviolet', 
                'mediumseagreen', 
                'orangered',
                'darkturquoise',
                'goldenrod'],
        wedgeprops=dict(width=0.4, edgecolor='white'),
        autopct='%.2f%%', textprops={'fontsize': 13.5},
        pctdistance=0.69)

plt.legend(labelsInside, loc=(0.85, 0.7), fontsize = 15)

plt.title('Nord Carolina', fontsize = 25)
plt.show()
```
