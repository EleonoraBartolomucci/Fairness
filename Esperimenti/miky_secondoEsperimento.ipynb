```
from pydrive.auth import GoogleAuth
from pydrive.drive import GoogleDrive
from oauth2client.client import GoogleCredentials
from google.colab import auth
import pandas as pd
import csv
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

auth.authenticate_user()
gauth = GoogleAuth()
gauth.credentials = GoogleCredentials.get_application_default()
drive = GoogleDrive(gauth)

sort = 'Random'

# file con i dati di tutti i ristoranti random
all_analysis = '1n9vdCnppJGdE6M6HdKYlksuTntQYHA7P'  # FILE ID, got on google drive with condivision link
download = drive.CreateFile({'id': all_analysis})
download.GetContentFile('AllRestaurant_' + '%s' % sort + '_Analysis.csv')

# leggo il csv con pandas
res = pd.read_csv('AllRestaurant_' + '%s' % sort + '_Analysis.csv')
```

```
import warnings
import numpy as np

# Prendo i dati che mi servono di tutti i ristoranti

positive = res['review_sentiment'] == 'positive'
negative = res['review_sentiment'] == 'negative'
neutral = res['review_sentiment'] == 'neutral'
feminine = res['gender'] == 'feminine'
masculine = res['gender'] == 'masculine'
american_restaurant = res['typology'] == 'American'
asian_restaurant = res['typology'] == 'Japanese'
mexican_restaurant = res['typology'] == 'Mexican'


# elimino dal dataframe le righe che non sono state riconosciute e presentano questa scritta
res = res[((res.age != 'Not recognized'))]

# sopprimo il Future Warning
warnings.simplefilter(action='ignore', category=FutureWarning)

# converto le stringhe di age in int
res.age = pd.to_numeric(res.age, errors='coerce').fillna(0).astype(np.int64)

under30 = res['age'] < 30
over30 = res['age'] > 30
under40 = res['age'] < 40
over40 = res['age'] > 40
under55 = res['age'] < 55
over55 = res['age'] > 55
```

```
# RISTORANTI AMERICANI

# women
totWomenUsers_under30_AmericanRestaurants = len(res[(american_restaurant) & (feminine)  & (under30)])  # women under 30
totWomenUsers_over30_and_under40_AmericanRestaurants = len(res[(american_restaurant) & (feminine)  & (over30) & (under40)])  # women between 30 and 40
totWomenUsers_over40_and_under55_AmericanRestaurants = len(res[(american_restaurant) & (feminine) & (over40) & (under55)])  # women between 40 and 55
totWomenUsers_over55_AmericanRestaurants = len(res[(american_restaurant) & (feminine)  & (over55)])  # women over 55

# men
totMenUsers_under30_AmericanRestaurants = len(res[(american_restaurant) & (masculine)  & (under30)])  # men under 30
totMenUsers_over30_and_under40_AmericanRestaurants = len(res[(american_restaurant) & (masculine)  & (over30) & (under40)])  # men between 30 and 40
totMenUsers_over40_and_under55_AmericanRestaurants = len(res[(american_restaurant) & (masculine) & (over40) & (under55)])  # men between 40 and 55
totMenUsers_over55_AmericanRestaurants = len(res[(american_restaurant) & (masculine)  & (over55)])  # men over 55


# POSITIVI

women_under30_americanrest_positive = (len(res[(american_restaurant) & (feminine) & (positive) & (under30)]) * 100) / totWomenUsers_under30_AmericanRestaurants
women_over30_and_under40_americanrest_positive = (len(res[(american_restaurant) & (feminine) & (positive) & (over30) & (under40)]) * 100) / totWomenUsers_over30_and_under40_AmericanRestaurants
women_over40_and_under55_americanrest_positive = (len(res[(american_restaurant) & (feminine) & (positive) & (over40) & (under55)]) * 100) / totWomenUsers_over40_and_under55_AmericanRestaurants
women_over55_americanrest_positive = (len(res[(american_restaurant) & (feminine) & (positive) & (over55)]) * 100) / totWomenUsers_over55_AmericanRestaurants

men_under30_americanrest_positive = (len(res[(american_restaurant) & (masculine) & (positive) & (under30)]) * 100) / totMenUsers_under30_AmericanRestaurants
men_over30_and_under40_americanrest_positive = (len(res[(american_restaurant) & (masculine) & (positive) & (over30) & (under40)]) * 100) / totMenUsers_over30_and_under40_AmericanRestaurants
men_over40_and_under55_americanrest_positive = (len(res[(american_restaurant) & (masculine) & (positive) & (over40) & (under55)]) * 100) / totMenUsers_over40_and_under55_AmericanRestaurants
men_over55_americanrest_positive = (len(res[(american_restaurant) & (masculine) & (positive) & (over55)]) * 100) / totMenUsers_over55_AmericanRestaurants


# NEGATIVI

women_under30_americanrest_negative = (len(res[(american_restaurant) & (feminine) & (negative) & (under30)]) * 100) / totWomenUsers_under30_AmericanRestaurants
women_over30_and_under40_americanrest_negative = (len(res[(american_restaurant) & (feminine) & (negative) & (over30) & (under40)]) * 100) / totWomenUsers_over30_and_under40_AmericanRestaurants
women_over40_and_under55_americanrest_negative = (len(res[(american_restaurant) & (feminine) & (negative) & (over40) & (under55)]) * 100) / totWomenUsers_over40_and_under55_AmericanRestaurants
women_over55_americanrest_negative = (len(res[(american_restaurant) & (feminine) & (negative) & (over55)]) * 100) / totWomenUsers_over55_AmericanRestaurants

men_under30_americanrest_negative = (len(res[(american_restaurant) & (masculine) & (negative) & (under30)]) * 100) / totMenUsers_under30_AmericanRestaurants
men_over30_and_under40_americanrest_negative = (len(res[(american_restaurant) & (masculine) & (negative) & (over30) & (under40)]) * 100) / totMenUsers_over30_and_under40_AmericanRestaurants
men_over40_and_under55_americanrest_negative = (len(res[(american_restaurant) & (masculine) & (negative) & (over40) & (under55)]) * 100) / totMenUsers_over40_and_under55_AmericanRestaurants
men_over55_americanrest_negative = (len(res[(american_restaurant) & (masculine) & (negative) & (over55)]) * 100) / totMenUsers_over55_AmericanRestaurants


# NEUTRI

women_under30_americanrest_neutral = (len(res[(american_restaurant) & (feminine) & (neutral) & (under30)]) * 100) / totWomenUsers_under30_AmericanRestaurants
women_over30_and_under40_americanrest_neutral = (len(res[(american_restaurant) & (feminine) & (neutral) & (over30) & (under40)]) * 100) / totWomenUsers_over30_and_under40_AmericanRestaurants
women_over40_and_under55_americanrest_neutral = (len(res[(american_restaurant) & (feminine) & (neutral) & (over40) & (under55)]) * 100) / totWomenUsers_over40_and_under55_AmericanRestaurants
women_over55_americanrest_neutral = (len(res[(american_restaurant) & (feminine) & (neutral) & (over55)]) * 100) / totWomenUsers_over55_AmericanRestaurants

men_under30_americanrest_neutral = (len(res[(american_restaurant) & (masculine) & (neutral) & (under30)]) * 100) / totMenUsers_under30_AmericanRestaurants
men_over30_and_under40_americanrest_neutral = (len(res[(american_restaurant) & (masculine) & (neutral) & (over30) & (under40)]) * 100) / totMenUsers_over30_and_under40_AmericanRestaurants
men_over40_and_under55_americanrest_neutral = (len(res[(american_restaurant) & (masculine) & (neutral) & (over40) & (under55)]) * 100) / totMenUsers_over40_and_under55_AmericanRestaurants
men_over55_americanrest_neutral = (len(res[(american_restaurant) & (masculine) & (neutral) & (over55)]) * 100) / totMenUsers_over55_AmericanRestaurants
```

```
# GRAFICO PER RISTORANTI AMERICANI

labels = ['Women Under30', 'Women Over30 and Under40', 'Women Over40 and Under55', 'Women Over55',
          'Men Under30', 'Men Over30 and Under40', 'Men Over40 and Under55', 'Men Over55']


americanPositive = [women_under30_americanrest_positive, women_over30_and_under40_americanrest_positive, women_over40_and_under55_americanrest_positive, women_over55_americanrest_positive,
                    men_under30_americanrest_positive, men_over30_and_under40_americanrest_positive, men_over40_and_under55_americanrest_positive, men_over55_americanrest_positive]

americanNegative = [women_under30_americanrest_negative, women_over30_and_under40_americanrest_negative, women_over40_and_under55_americanrest_negative, women_over55_americanrest_negative,
                    men_under30_americanrest_negative, men_over30_and_under40_americanrest_negative, men_over40_and_under55_americanrest_negative, men_over55_americanrest_negative]

americanNeutral = [women_under30_americanrest_neutral, women_over30_and_under40_americanrest_neutral, women_over40_and_under55_americanrest_neutral, women_over55_americanrest_neutral,
                    men_under30_americanrest_neutral, men_over30_and_under40_americanrest_neutral, men_over40_and_under55_americanrest_neutral, men_over55_americanrest_neutral]

print(women_under30_americanrest_positive + women_under30_americanrest_negative + women_under30_americanrest_neutral)
print(women_over30_and_under40_americanrest_positive + women_over30_and_under40_americanrest_negative + women_over30_and_under40_americanrest_neutral)
print(women_over40_and_under55_americanrest_positive + women_over40_and_under55_americanrest_negative + women_over40_and_under55_americanrest_neutral)
print(women_over55_americanrest_positive + women_over55_americanrest_negative + women_over55_americanrest_neutral)

print(men_under30_americanrest_positive + men_under30_americanrest_negative + men_under30_americanrest_neutral)
print(men_over30_and_under40_americanrest_positive + men_over30_and_under40_americanrest_negative + men_over30_and_under40_americanrest_neutral)
print(men_over40_and_under55_americanrest_positive + men_over40_and_under55_americanrest_negative + men_over40_and_under55_americanrest_neutral)
print(men_over55_americanrest_positive + men_over55_americanrest_negative + men_over55_americanrest_neutral)


x = np.arange(len(labels))  # the label locations
width = 0.15  # the width of the bars

matplotlib.rcParams['figure.figsize'] = (18.0, 10.0)

fig, ax = plt.subplots()
rects1 = ax.bar(x - width, americanPositive, width, label='positive')
rects2 = ax.bar(x, americanNegative, width, label='negative')
rects3 = ax.bar(x + width, americanNeutral, width, label='neutral')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('American Restaurants')
ax.set_title('Sentiment analysis by group of users')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{:.2%}'.format(height/100),
                    xy=(rect.get_x() + rect.get_width(), height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')

autolabel(rects1)
autolabel(rects2)
autolabel(rects3)

fig.tight_layout()

plt.show()
```
