```
from pydrive.auth import GoogleAuth
from pydrive.drive import GoogleDrive
from oauth2client.client import GoogleCredentials
from google.colab import auth
import pandas as pd
import csv
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

auth.authenticate_user()
gauth = GoogleAuth()
gauth.credentials = GoogleCredentials.get_application_default()
drive = GoogleDrive(gauth)

sort = 'Random'

# file con i dati di tutti i ristoranti random
all_analysis = '1n9vdCnppJGdE6M6HdKYlksuTntQYHA7P'  # FILE ID, got on google drive with condivision link
download = drive.CreateFile({'id': all_analysis})
download.GetContentFile('AllRestaurant_' + '%s' % sort + '_Analysis.csv')

# leggo il csv con pandas
res = pd.read_csv('AllRestaurant_' + '%s' % sort + '_Analysis.csv')
```

```
# Prendo i dati che mi servono dei ristoranti

asian = res['ethnicity'] == 'asian'
mexican = res['ethnicity'] == 'hispanic, latino, or spanish origin'
black_american = res['ethnicity'] == 'black or african american'
white_american = res['ethnicity'] == 'white'
positive = res['review_sentiment'] == 'positive'
negative = res['review_sentiment'] == 'negative'
neutral = res['review_sentiment'] == 'neutral'
american_restaurant = res['typology'] == 'American'
asian_restaurant = res['typology'] == 'Japanese'
mexican_restaurant = res['typology'] == 'Mexican'
nc = res['zone'] == 'NC'  # North Carolina
az = res['zone'] == 'AZ'  # Arizona
nv = res['zone'] == 'NV'  # Nevada

# elimino dal dataframe le righe che non sono state riconosciute e presentano questa scritta
res = res[((res.age != 'Not recognized'))]

# AMERICAN RESTAURANTS

totNorthCarolinaUsers_AmericanRestaurants = len(res[(american_restaurant) & (nc)])
totArizonaUsers_AmericanRestaurants = len(res[(american_restaurant) & (az)])
totNevadaUsers_AmericanRestaurants = len(res[(american_restaurant) & (nv)])

# POSITIVI
# numero medio di positive sentiment di utenti asiatici, messicani e americani sul totale della rispettiva etnia
nc_american_positive = (len(res[(american_restaurant) & (nc) & (positive)]) * 100) / totNorthCarolinaUsers_AmericanRestaurants
az_american_positive = (len(res[(american_restaurant) & (az) & (positive)]) * 100) / totArizonaUsers_AmericanRestaurants
nv_american_positive = ((len(res[(american_restaurant) & (nv) & (positive)])) * 100) / totNevadaUsers_AmericanRestaurants
                          
# NEGATIVI
# numero medio di negative sentiment di utenti asiatici, messicani e americani sul totale della rispettiva etnia
nc_american_negative = (len(res[(american_restaurant) & (nc) & (negative)]) * 100) / totNorthCarolinaUsers_AmericanRestaurants
az_american_negative = (len(res[(american_restaurant) & (az) & (negative)]) * 100) / totArizonaUsers_AmericanRestaurants
nv_american_negative = ((len(res[(american_restaurant) & (nv) & (negative)])) * 100) / totNevadaUsers_AmericanRestaurants

# NEUTRI
# numero medio di neutral sentiment di utenti asiatici, messicani e americani sul totale della rispettiva etnia
nc_american_neutral = (len(res[(american_restaurant) & (nc) & (neutral)]) * 100) / totNorthCarolinaUsers_AmericanRestaurants
az_american_neutral = (len(res[(american_restaurant) & (az) & (neutral)]) * 100) / totArizonaUsers_AmericanRestaurants
nv_american_neutral = ((len(res[(american_restaurant) & (nv) & (neutral)])) * 100) / totNevadaUsers_AmericanRestaurants

# -----------------------------------------------------------------------------------------------------------------------------------

# ASIAN RESTAURANTS

totNorthCarolinaUsers_AsianRestaurants = len(res[(asian_restaurant) & (nc)])
totArizonaUsers_AsianRestaurants = len(res[(asian_restaurant) & (az)])
totNevadaUsers_AsianRestaurants = len(res[(asian_restaurant) & (nv)])

# POSITIVI
# numero medio di positive sentiment di utenti del north carolina, arizona e nevada
nc_asian_positive = (len(res[(asian_restaurant) & (nc) & (positive)]) * 100) / totNorthCarolinaUsers_AsianRestaurants
az_asian_positive = (len(res[(asian_restaurant) & (az) & (positive)]) * 100) / totArizonaUsers_AsianRestaurants
nv_asian_positive = ((len(res[(asian_restaurant) & (nv) & (positive)])) * 100) / totNevadaUsers_AsianRestaurants
                          
# NEGATIVI
# numero medio di positive sentiment di utenti del north carolina, arizona e nevada
nc_asian_negative = (len(res[(asian_restaurant) & (nc) & (negative)]) * 100) / totNorthCarolinaUsers_AsianRestaurants
az_asian_negative = (len(res[(asian_restaurant) & (az) & (negative)]) * 100) / totArizonaUsers_AsianRestaurants
nv_asian_negative = ((len(res[(asian_restaurant) & (nv) & (negative)])) * 100) / totNevadaUsers_AsianRestaurants

# NEUTRI
# numero medio di neutral sentiment di utenti asiatici, messicani e americani sul totale della rispettiva etnia
nc_asian_neutral = (len(res[(asian_restaurant) & (nc) & (neutral)]) * 100) / totNorthCarolinaUsers_AsianRestaurants
az_asian_neutral = (len(res[(asian_restaurant) & (az) & (neutral)]) * 100) / totArizonaUsers_AsianRestaurants
nv_asian_neutral = ((len(res[(asian_restaurant) & (nv) & (neutral)])) * 100) / totNevadaUsers_AsianRestaurants

# -----------------------------------------------------------------------------------------------------------------------------------

# MEXICAN RESTAURANTS

totNorthCarolinaUsers_MexicanRestaurants = len(res[(mexican_restaurant) & (nc)])
totArizonaUsers_MexicanRestaurants = len(res[(mexican_restaurant) & (az)])
totNevadaUsers_MexicanRestaurants = len(res[(mexican_restaurant) & (nv)])

# POSITIVI
# numero medio di positive sentiment di utenti del north carolina, arizona e nevada
nc_mexican_positive = (len(res[(mexican_restaurant) & (nc) & (positive)]) * 100) / totNorthCarolinaUsers_MexicanRestaurants
az_mexican_positive = (len(res[(mexican_restaurant) & (az) & (positive)]) * 100) / totArizonaUsers_MexicanRestaurants
nv_mexican_positive = ((len(res[(mexican_restaurant) & (nv) & (positive)])) * 100) / totNevadaUsers_MexicanRestaurants
                          
# NEGATIVI
# numero medio di positive sentiment di utenti del north carolina, arizona e nevada
nc_mexican_negative = (len(res[(mexican_restaurant) & (nc) & (negative)]) * 100) / totNorthCarolinaUsers_MexicanRestaurants
az_mexican_negative = (len(res[(mexican_restaurant) & (az) & (negative)]) * 100) / totArizonaUsers_MexicanRestaurants
nv_mexican_negative = ((len(res[(mexican_restaurant) & (nv) & (negative)])) * 100) / totNevadaUsers_MexicanRestaurants

# NEUTRI
# numero medio di neutral sentiment di utenti asiatici, messicani e americani sul totale della rispettiva etnia
nc_mexican_neutral = (len(res[(mexican_restaurant) & (nc) & (neutral)]) * 100) / totNorthCarolinaUsers_MexicanRestaurants
az_mexican_neutral = (len(res[(mexican_restaurant) & (az) & (neutral)]) * 100) / totArizonaUsers_MexicanRestaurants
nv_mexican_neutral = ((len(res[(mexican_restaurant) & (nv) & (neutral)])) * 100) / totNevadaUsers_MexicanRestaurants
```

```
# Costruisco il grafico

labels = ['NorthCarolinaSentiment', 'ArizonaSentiment', 'NevadaSentiment']

# GRAFICO PER I RISTORANTI PER ZONA -> CHI E' CHE HA PIU RECENSIONI POSITIVE, NEGATIVE E NEUTRE TRA I TRE RISTORANTI AMERICANI

# valori barre
positive = [nc_american_positive, az_american_positive, nv_american_positive]
negative = [nc_american_negative, az_american_negative, nv_american_negative]
neutral = [nc_american_neutral, az_american_neutral, nv_american_neutral]

print(nc_american_positive + nc_american_negative + nc_american_neutral)
print(az_american_positive + az_american_negative + az_american_neutral)
print(nv_american_positive + nv_american_negative + nv_american_neutral)

x = np.arange(len(labels))  # the label locations
width = 0.15  # the width of the bars

matplotlib.rcParams['figure.figsize'] = (10.0, 6.0)
fig, ax = plt.subplots()

rects1 = ax.bar(x - width, positive, width, label='positive')
rects2 = ax.bar(x, negative, width, label='negative', color='red')
rects3 = ax.bar(x + width, neutral, width, label='neutral', color='orange')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('All American Restaurants')
ax.set_title('General yelp users sentiment')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{:.2%}'.format(height/100),
                    xy=(rect.get_x() + rect.get_width(), height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')

autolabel(rects1)
autolabel(rects2)
autolabel(rects3)

fig.tight_layout()

plt.show()
```
